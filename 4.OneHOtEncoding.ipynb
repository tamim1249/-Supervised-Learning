One Hot Encoding কী?

যখন আমাদের কাছে categorical (শ্রেণীভিত্তিক) ডেটা থাকে, যেমন রঙ (লাল, নীল, সবুজ), দেশের নাম (বাংলাদেশ, ভারত, পাকিস্তান), তাহলে সরাসরি সেই নামগুলোকে মডেলে ফিড করা সম্ভব না। কারণ মেশিন লার্নিং মডেল সংখ্যা বুঝে, আর এই ক্যাটেগরি মানগুলো সংখ্যার মতো দেখতে হলেও তারা আসলে নাম (string), তাই একটা কৌশল দরকার।

One Hot Encoding হলো সেই কৌশল যেখানে প্রতিটি ক্যাটেগরিকে আলাদা একটি binary (0/1) কলামে রূপান্তর করা হয়।

কেন দরকার?

    যদি তুমি শুধুমাত্র নামগুলোকে সংখ্যায় রূপান্তর করো যেমন আম=1, কলা=2, আপেল=3, তাহলে মডেল মনে করবে কলা (2) আপেলের (3) থেকে ছোট কিন্তু আম (1) উভয়ের থেকেও ছোট, অর্থাৎ একটা অর্ডার আছে। কিন্তু আসলে এটা ভুল — আম, কলা, আপেল কোনটারই বেশি বা কম না, শুধুমাত্র আলাদা আলাদা শ্রেণী।

    One Hot Encoding এই ভুল বুঝাবুঝি থেকে রক্ষা করে।

এক কথায়:

    One Hot Encoding categorical variables কে numerical vector এ রূপান্তর করার পদ্ধতি, যেখানে প্রতিটি ক্যাটেগরি আলাদা একটি কলাম হিসেবে থাকে।

    প্রত্যেক রেকর্ডে শুধুমাত্র সেই ক্যাটেগরির কলামে ১ থাকবে, বাকিগুলো ০।



```python
import pandas as pd
df=pd.read_csv('homeprices.csv')
df
```

```python
d=pd.get_dummies(df.town)
d
```

```python
marge=pd.concat([df,d],axis='columns')
marge
```

```python
final=marge.drop(['town','west windsor'],axis='columns')
final
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
X=final.drop(['price'],axis='columns')
X
```

```python
y=final.price
y
```

```python
model.fit(X,y)
model.predict([[3600,0,0]])
```

```python
model.score(X,y)
```

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

# Sample DataFrame
df = pd.DataFrame({
    'Department': ['Sales', 'HR', 'IT', 'Sales', 'IT'],
    'Education': ['Bachelors', 'Masters', 'Bachelors', 'PhD', 'Masters'],
    'Gender': ['Male', 'Female', 'Female', 'Male', 'Male'],
    'Experience': [3, 7, 2, 5, 10],
    'Salary': [40000, 50000, 42000, 60000, 75000]
})

# Step 1: Select categorical columns only
cat_cols = ['Department', 'Education', 'Gender']

# Step 2: Create encoder and fit_transform
encoder = OneHotEncoder(drop='first', sparse_output=False)
encoded_data = encoder.fit_transform(df[cat_cols])

# Step 3: Get column names
column_names = encoder.get_feature_names_out(cat_cols)

# Step 4: Create encoded DataFrame
encoded_df = pd.DataFrame(encoded_data, columns=column_names)

# Step 5: Combine with numeric columns
final_df = pd.concat([encoded_df, df[['Experience', 'Salary']]], axis=1)

# Step 6: Show result
print(final_df)
```
